```python
## 7. What contributed to the survival of those who survived?
#Check if there are missing values
df.isnull().sum()
```

```python
#Clean data and drop unnecessary columns
df = df.drop(["PassengerId","Ticket","Cabin"], axis = 1)
df
```

```python
#fill in missing value for age based on mean age value in regards to standard deviation
for dataset in df:
    mean = df["Age"].mean()
    std = df["Age"].std()
    is_null = df["Age"].isnull().sum()
    
    rand_age = np.random.randint(mean - std, mean + std, size = is_null)
    
    age_slice = df["Age"].copy()
    age_slice[np.isnan(age_slice)] = rand_age
    df["Age"] = age_slice
    df["Age"] = df["Age"].astype(int)
df["Age"].isnull().sum()
```

```python
df['Embarked'].describe()
```

```python
#replace missing values with the most common one 'S'
df['Embarked'] = df['Embarked'].replace(np.nan, 'S')
```

```python
df.info()
```

```python
#Converting “Fare” from float to int64]
df['Fare'] = df['Fare'].astype(int)
```

```python
## One hot encode Sex and Embarked column
df = pd.get_dummies(df, columns=["Sex","Embarked"])
```

```python
df
```

```python
import re
# Extract title from name.
def get_title(name):
    title_search = re.search(' ([A-Za-z]+)\.', name)
    if title_search:
        return title_search.group(1)

titles = df["Name"].apply(get_title)
title_mapping = {"Mr": 1, "Ms": 2, "Miss": 2, "Mrs": 3, "Master": 4, "Dr": 5, "Rev": 6, "Major": 7, "Col": 7, "Capt": 7, "Mlle": 8, "Mme": 8, "Don": 9, "Sir": 9, "Lady": 10, "Countess": 10, "Jonkheer": 10}
#dataset was created using the titles as keys and the counts as the values
for k,v in title_mapping.items():
    titles[titles == k] = v
```

```python
df.insert(2,'Titles',titles)
df
```

```python
#create#categories for age
category=pd.cut(df['Age'],bins=[-1,15,20,25,32,40,100], labels=['0','1','2','3','4','5'])
df.insert(5,'Age Cat',category)
df
```

```python
df['Age Cat'].value_counts()
```

```python
#create categories for fare
categoryf=pd.cut(df['Fare'],bins=[-1,7,15,30,60,1000], labels=['0','1','2','3','4'])
df.insert(9,'Fare Cat',categoryf)
df
```

```python
df['Fare Cat'].value_counts()
```

```python
df = df.drop(columns=['Name','Age','Fare'])
df
```

```python
X=df.drop(columns=['Survived'])
Y=df["Survived"]
```

```python
# Split the dataset into Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

```python
X_train.isnull().sum()
```

```python
#Random forest
from sklearn.ensemble import RandomForestClassifier
classifier_rf= RandomForestClassifier(n_estimators = 100, criterion = 'entropy', random_state = 0)
classifier_rf.fit(X_train, y_train)

# Predicting the Test set results
y_pred_rf= classifier_rf.predict(X_test)
```

```python
from sklearn.metrics import roc_auc_score
auc_rf = roc_auc_score(y_test, y_pred_rf)
print('AUC score =%.3f' % (auc_rf))
```

```python
importances_rf = pd.DataFrame(classifier_rf.feature_importances_,index = X_train.columns,columns=['importance']).sort_values('importance',ascending=False)
print(importances_rf)
#Age cat has the highest importance, contributes the most to the survival
```
